```python
#import csv file
df = pd.read_csv("base_etablissement_par_tranche_effectif.csv")
```

```python
df.sample(10)
```

```python
df["ALL_FIRMS"] = df["E14TST"] + df["E14TS0ND"] + df["E14TS1"] + df["E14TS6"] + df["E14TS10"] + df["E14TS20"] + df["E14TS50"] + df["E14TS100"] + df["E14TS200"] + df["E14TS500"]
```

```python
df.sample(10)
```

```python
reg2 = df[["CODGEO", "REG", "ALL_FIRMS"]]
```

```python
reg = df[["REG", "ALL_FIRMS"]]
```

```python
group = reg.groupby("REG", as_index=False).sum()
group
```

```python
#import csv file
ng = pd.read_csv("name_geographic_information.csv")
```

```python
ng.columns
```

```python
ng["code_insee"]
```

```python
ng2 = ng[["code_insee", "REG"]]
ng2.rename(columns={'code_insee': 'CODGEO'}, inplace=True)
ng2
```

```python
ng2["code_insee"].
```

```python
ng.rename(columns={'code_région': 'REG', 'nom_région': 'NAME_REG'}, inplace=True)
```

```python
regions = ng[["REG", "NAME_REG"]]
regions = regions.drop_duplicates()
regions
```

```python
result = pd.merge(group, regions, on='REG', how='outer')
result
```

```python
result = result[result['ALL_FIRMS'].notna()]
```

```python
result = result.sort_values("ALL_FIRMS", ascending=False)
```

```python
palette = sns.color_palette("Blues_r", 28)
```

```python
sns.set_style("whitegrid")
ax = sns.barplot(result, x="NAME_REG", y="ALL_FIRMS", palette=palette)
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain', axis='y')
ax.set(xlabel='region name', ylabel='number of companies')

pass
```

```python
#import csv file
salary = pd.read_csv("net_salary_per_town_categories.csv")
```

```python
salary.sample(10)
```

```python
sal = salary[["CODGEO", "SNHM14"]]
```

```python
sal.rename(columns={'SNHM14': 'MEAN_SALARY'}, inplace=True)
sal
```

```python
reg2
```

```python
salary2 = pd.merge(sal, reg2, on='CODGEO', how='outer')
salary2
```

```python
salary2 = salary2[salary2['MEAN_SALARY'].notna()]
salary2
```

```python
sal_by_reg = salary2[["REG", "MEAN_SALARY"]]
sal_by_reg = sal_by_reg.groupby("REG", as_index=False).mean()
sal_by_reg["MEAN_SALARY"] = round(sal_by_reg["MEAN_SALARY"], 2)
sal_by_reg
```

```python
result
```

```python
final = pd.merge(result, sal_by_reg, on='REG', how='outer')
final
```

```python
firms_to_salary = final['ALL_FIRMS'].corr(final['MEAN_SALARY'])
firms_to_salary
```

```python
final.to_csv('sal_firms.csv', sep='\t', encoding='utf-8')
```

```python
sns.set_style("whitegrid")
ax = sns.barplot(data = final, x="NAME_REG", y="MEAN_SALARY", palette=palette)
plt.xticks(rotation=90)
ax.set(xlabel='region name', ylabel='mean salary ( €/h )')


pass
```
